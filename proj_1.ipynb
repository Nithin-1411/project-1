```python
import pandas as pd
import numpy as np

# Assuming you have a DataFrame df with your dataset and labels
# Assuming your labels are in a column named 'y'
# Assuming your features are in the remaining columns

unique_classes = df['class_name'].unique()
print(unique_classes)
class_centroids = [df[df['class_name'] == c].mean(axis=0) for c in unique_classes]
class_spreads = [df[df['class_name'] == c].std(axis=0) for c in unique_classes]
print(class_centroids)
# Calculate the distance between mean vectors between classes
distances_between_classes = []
for i in range(len(unique_classes)):
    for j in range(i + 1, len(unique_classes)):
        distance = np.linalg.norm(class_centroids[i] - class_centroids[j])
        distances_between_classes.append((unique_classes[i], unique_classes[j], distance))

# distances_between_classes now contains the interclass distances
```
